```python
# F. Plotting multiple Matplotlib and Seaborn charts in a grid 
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))

# Plotting
axes[0,0].plot(years, cycle, 's-b')
axes[0,0].plot(years, bike, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield')
axes[0,0].legend(['Cycle', 'Bike']);
axes[0,0].set_title('Yields in Ukreine')

# Scatter plot
axes[0,1].set_title('Sepal Length & Sepal Width')
sns.scatterplot(x=iris_df.sepal_length, y=iris_df.sepal_width, 
                hue=iris_df.variety, s=50, ax=axes[0,1]);

# Histrofram
axes[1,0].set_title('Distribution of Sepal')
axes[1,0].hist([iris_df.sepal_width, iris_df.sepal_width, iris_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);
axes[1,0].legend(['Setosa', 'Versicolor', 'Virginica']);

# Heatmap
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Greens', ax=axes[1,1]);

plt.tight_layout(pad=5);
```
